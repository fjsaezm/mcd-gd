```python
import pandas as pd

file_name = 'uam-personal-pdi-2019-anonimizado.csv'
```

```python
df = pd.read_csv(file_name)

df = df.drop(columns=['IMPORTANTE'])
df = df.loc[:, ~df.columns.str.startswith('cod_')]
df = df.loc[:, ~df.columns.str.startswith('lat_')]
df = df.loc[:, ~df.columns.str.startswith('lon_')]

print(df.shape)
df.head()
```

```python
# Count the times of unique value for row des_unidad_responsable appeared
df.groupby('des_unidad_responsable').count()
```

```python
# Limit our study using the pivot variables
df_limited = df.loc[(df['des_unidad_responsable'] == 'Departamento de Ingeniería Informática') &
                    (df['des_genero'] == 'Hombre')]
df_limited.shape
```

# Coherence block 1

```python
# Coherence block 1
df_limited.filter(['des_universidad','anio'], axis=1).groupby(['des_universidad','anio']).size()
```

# Coherence block 2

```python
# Coherence block 2
n_rows = df_limited.shape[0]
variables = ['des_pais_nacionalidad',
             'des_continente_nacionalidad',
             'des_agregacion_paises_nacionalidad']

def obtain_freq(variables, data=df_limited):
    n_rows = data.shape[0]
    return pd.DataFrame( 
        data.filter(variables, axis=1).groupby(variables, dropna=False).size() / n_rows )

obtain_freq(variables)
```

# Coherence block 3

```python
# Coherence block 3
variables = ['des_comunidad_residencia',
            'des_provincia_residencia',
            'des_municipio_residencia']

obtain_freq(variables)
```

# Coherence block 4

```python
# Coherence block 4
variables = ['anio_nacimiento']
df_limited_born = df_limited.loc[df_limited['anio_nacimiento'] < 1975]
obtain_freq(variables, data=df_limited_born)
```

# Coherence block 5

```python
# Coherence block 5
variables = [
'des_tipo_personal',
'des_categoria_cuerpo_escala',
'des_tipo_contrato',
'des_dedicacion',
'num_horas_semanales_tiempo_parcial',
'des_situacion_administrativa'
]

df_limited_prof = df_limited.loc[
    df_limited['des_categoria_cuerpo_escala'] == 'Profesor Titular de Universidad'
]
obtain_freq(variables, data=df_limited_prof)
```

# Coherence block 6

```python
# Coherence block 6
variables = ['ind_cargo_remunerado']
obtain_freq(variables)
```

# Coherence block 7

```python
# Coherence block 7
variables = [
'des_titulo_doctorado',
'des_pais_doctorado',
'des_continente_doctorado',
'des_agregacion_paises_doctorado',
'des_universidad_doctorado',
'anio_lectura_tesis',
'anio_expedicion_titulo_doctor',
'des_mencion_europea']

df_limited_thesis = df_limited.loc[(df_limited['anio_lectura_tesis'] == 2000) &
                           (df_limited['des_universidad_doctorado'] == 'Universidad Autónoma de Madrid')]

obtain_freq(variables, data=df_limited_thesis)
```

# Coherence block 8

```python
# Coherence block 8
variables = [
'des_tipo_unidad_responsable',
'des_area_conocimiento']

obtain_freq(variables)
```
